# Data Cleaning

```python
all_df = all_df[all_df['Order Date'].str[0:2] != 'Or']
all_df.head()
```

```python
all_df['Month'] = all_df['Order Date'].str[0:2]
all_df['Month'] = all_df['Month'].astype('int32')
all_df.head()
```

```python
all_df['Price Each'] = all_df['Price Each'].astype('float32')
all_df['Quantity Ordered'] = all_df['Quantity Ordered'].astype('int32')
```

```python
all_df['Sales'] = all_df['Quantity Ordered'] * all_df['Price Each']
```

```python
all_df.head()
```

# What is the best month for sale ? How much did it sell ?

```python
all_df.groupby('Month').sum()
results = all_df.groupby('Month').sum()
results
```

```python
months = range(1,13)
plt.bar(months, results['Sales'])
plt.ylabel('Rupees')
plt.xlabel('Months')
plt.show()
```

# Which city had the highest number of sales ?

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_df['City'] = all_df['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ')')
```

```python
all_df.head()
```

```python
results = all_df.groupby('City').sum()
results
```

```python
city = [city for city, df in all_df.groupby('City')]

plt.bar(city, results['Sales'])
plt.rcParams["figure.figsize"] = (20,6)
plt.ylabel('Rupees')
plt.xlabel('City')
```

# What products are most often sold together ?

```python
df = all_df[all_df['Order ID'].duplicated(keep = False)]


df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))


df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(20)
```

```python
from itertools import combinations
from collections import Counter
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
print(count)
```

```python
for (key, value) in count.most_common(20):
    print(key, value)
```

# What product sold the most ? Give the reason

```python
product_group = all_df.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
quantity_ordered
```

```python
products = [product for product, df in product_group]
plt.bar(products, quantity_ordered)
plt.xticks(products, rotation='vertical', size = 14)
plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
prices = all_df.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color = 'g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Products name')
ax1.set_ylabel('Quantity Ordered', color = 'g')

ax2.set_ylabel('Price ($)', color = 'b')
ax1.set_xticklabels(products, rotation='vertical', size = 14)
plt.show()
```
